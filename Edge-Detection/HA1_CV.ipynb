```
north_k = np.array([[-3,-3,5], #North and West kirsch Kernels
           [-3,0,5],
           [-3,-3,5]])
      
west_k = np.array([[5,5,5],
          [-3,0,-3],
          [-3,-3,-3]])

print(north_k)
print(west_k)
```

```
def convol(kernel, img):   #Function for convolution of image with kernel
  con = []
  height, width = img.shape
  for i in range(1,height-1):
    for j in range(1,width-1):
      con.append(sum(sum(kernel*img[i-1:i+2,j-1:j+2])))
      #print(con)
  conv = np.array(con)
  conv=conv.reshape(img.shape[0]-2, img.shape[1]-2)
  return conv
```

```
north = convol(north_k, arr)
west = convol(west_k, arr)
cv2_imshow(north)
cv2_imshow(west)
```
